# Plotting Example With Experimental Aggregations

Just a heads up - this notebook uses upcoming experimental features which _may_ be subject to change in the future.

```python
import sage_data_client
import matplotlib.pyplot as plt
```

First, we'll query the daily min / mean / max aggregations over the last 90 days of temperature from W023's BME680 sensor.

```python
df_min = sage_data_client.query(
    start="-90d",
    filter={
        "name": "env.temperature",
        "vsn": "W023",
        "sensor": "bme680",
    },
    experimental_func="min",
    experimental_window="1d",
)

df_mean = sage_data_client.query(
    start="-90d",
    filter={
        "name": "env.temperature",
        "vsn": "W023",
        "sensor": "bme680",
    },
    experimental_func="mean",
    experimental_window="1d",
)

df_max = sage_data_client.query(
    start="-90d",
    filter={
        "name": "env.temperature",
        "vsn": "W023",
        "sensor": "bme680",
    },
    experimental_func="max",
    experimental_window="1d",
)
```

Next, we'll plot a line chart with all the values together.

```python
plt.figure()
plt.plot(df_min["timestamp"], df_min["value"], label="min")
plt.plot(df_mean["timestamp"], df_mean["value"], label="mean")
plt.plot(df_max["timestamp"], df_max["value"], label="max")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
